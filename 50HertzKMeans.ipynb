### Scale Data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(df_monthly)
scaled_X
```

### KMeans with K=2

```python
from sklearn.cluster import KMeans
# help(KMeans)
model = KMeans(n_clusters=2)
# Fit and predict in one step:
cluster_labels = model.fit_predict(scaled_X)
cluster_labels
```

```python
df_monthly['Cluster'] = cluster_labels
df_monthly.corr()
```

```python
# How the clustering correlate to other features:
df_monthly.corr()['Cluster'].sort_values()
# Looks like "2019-12" is the most strongly correlated and "2020-07" is most negatively correlated.
```

```python
# To see it in a bar chart:
plt.figure(figsize=(12,6), dpi=200)
df_monthly.corr()['Cluster'].iloc[:-1].sort_values().plot(kind='bar')
```

### Find the Optimal Value of K:

```python
# Elbow method:
# Note that the result of this method is never 100% correct.
ssd = []

for k in range(2, 10):
    model = KMeans(n_clusters=k)
    model.fit(scaled_X)
    # Append SSD point to SSD center distance:
    ssd.append(model.inertia_)
    
ssd
```

```python
plt.plot(range(2,10), ssd, 'o--')
```

```python
# The slope decreased at point 6. So for future researches, we will set n_clusters = 6.
```

### KMeans with K=6

```python
model_k6 = KMeans(n_clusters=6)
# Fit and predict in one step:
cluster_labels = model_k6.fit_predict(scaled_X)
cluster_labels
```

```python
df_monthly['Cluster'] = cluster_labels
df_monthly.corr()
```
